```python
import pandas as pd
from geopy import distance as dis
import itertools
import numpy as np
import re
import fuzzywuzzy
from fuzzywuzzy import process
from fuzzywuzzy import fuzz
```

```python
def calculate_distance(p1,p2):
    return dis.distance(p1,p2).km
```

```python
socail_media_df=pd.read_csv('social_media_Netherlands.csv',delimiter="|",index_col=0)
```

```python
socail_media_df
```

```python
web_scrapying_df=pd.read_csv("nl_result_full_name.csv",index_col=0)
```

```python
web_scrapying_df_location=web_scrapying_df[['lat','lng']]
```

```python
web_scrapying_df_location
```

```python
web_scrapying_df_location=web_scrapying_df_location.drop_duplicates()
```

```python
result_list=[]
for row in socail_media_df.iterrows():
    position=row[1][["latitude","longitude"]].to_list()
    distance_list=[]
    for candidate in web_scrapying_df_location.iterrows():
        distance_=calculate_distance(position,candidate[1].to_list())
        distance_list.append([distance_]+candidate[1].to_list())
    distance_list.sort(key=lambda x:x[0])
    count=sum([1 for i in distance_list if i[0]<0.1])
    
    if count==1:
        print(row[1])
        print("One candidate")
        print(distance_list[0])
        
    
    elif count==0:
        print(row[1])
        print("No candidate in suitable distance, print the most close candidate")
        print(distance_list[0])
    tmp=row[1]
    tmp['min_distance_web',]=distance_list[0][0]
    tmp['corrected_lat_web']=distance_list[0][1]
    tmp['corrected_lon_web']=distance_list[0][2]
    result_list.append(tmp)
```

```python
result_df=pd.concat(result_list,axis=1).T
```

```python
for threshold in np.arange(0.0,10,0.5):
    print(str(sum(result_df['min_distance']<=threshold))+" of addjustion with distance within "+str(threshold)+' KM')
```

```python
Gmap_df=pd.read_csv('GMaps_Netherlands.csv',index_col=0)
```

```python
Gmap_df_location=Gmap_df[["place_lat","place_long"]]
```

```python
Gmap_set=set([tuple(row[1].to_list()) for row in Gmap_df_location.drop_duplicates().iterrows()])
```

```python
previous_len=None
tmp_set=Gmap_set
while(previous_len!=len(tmp_set)):
    previous_len=len(tmp_set)
    flag=0
    print('---')
    for i in tmp_set:
        #print('--')
        
        for j in tmp_set:
            #print('--')
            if i!=j and calculate_distance(i,j)<1:
                #print('--')
                print(calculate_distance(i,j))
                k=((i[0]+j[0])/2,(i[1]+j[1])/2)
                tmp_set.remove(i)
                tmp_set.remove(j)
                tmp_set.add(k)
                flag=1
            if flag:
                break
        if flag:
            break
```

```python
merged_Gmap=pd.DataFrame(list(tmp_set))
merged_Gmap.columns=['lat','lon']
```

```python
result_list=[]
for row in result_df.iterrows():
    position=row[1][["latitude","longitude"]].to_list()
    distance_list=[]
    for candidate in merged_Gmap.iterrows():
        distance_=calculate_distance(position,candidate[1].to_list())
        distance_list.append([distance_]+candidate[1].to_list())
    distance_list.sort(key=lambda x:x[0])
    count=sum([1 for i in distance_list if i[0]<0.1])
    
    if count==1:
        print(row[1])
        print("One candidate")
        print(distance_list[0])
        
    
    elif count==0:
        print(row[1])
        print("No candidate in suitable distance, print the most close candidate")
        print(distance_list[0])
    tmp=row[1]
    tmp['min_distance_GMap',]=distance_list[0][0]
    tmp['corrected_lat_GMap']=distance_list[0][1]
    tmp['corrected_lon_GMap']=distance_list[0][2]
    result_list.append(tmp)
```

```python
result_df=pd.concat(result_list,axis=1).T
```

```python
result_df['min_distance']=result_df[['min_distance_web',"min_distance_GMap"]].min(axis=1)
```

```python
sm_gt=pd.read_csv("sm_gt.csv")
all_infor_gt=pd.read_csv('page_info_Guatemala.csv',delimiter='|',index_col=0)
infor_gt=pd.read_csv('pages_Guatemala_validate.csv',delimiter='|',index_col=0)
matched_df_gt=pd.read_csv('address_name_match_gt.csv')
```

```python
infor_gt=infor_gt[['fb_id','latitude','longitude']].merge(all_infor_gt,left_on='fb_id',how='left',right_on="id")
```

```python
name_list_gt=list(map(lambda x: x[1]['username'] if type(x[1]['username'])==str else x[1]['name'],infor_gt.iterrows()))
```

```python
name_list_gt_space=list(map(lambda x :re.sub("([a-z])([A-Z])","\g<1> \g<2>",x),name_list_gt))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 1))
```

```python
vectorizer.fit(name_list_gt_space)
```

```python
def top1(string):
    df=pd.DataFrame([vectorizer.get_feature_names(),vectorizer.transform([string]).todense()[0].tolist()[0]]).T.sort_values(1,ascending=False).head(1)
    df.columns=['topic','score']
    df['string']=string
    tmp_list=df.reset_index(drop=True).values.tolist()
    return tmp_list[0]
```

```python
def top1(string):
    df=pd.DataFrame([vectorizer.get_feature_names(),vectorizer.transform([string]).todense()[0].tolist()[0]]).T.sort_values(1,ascending=False)
    df=df[df[1]==df[1].max()]
    df.columns=['topic','score']
    df['string']=string
    tmp_list=df.reset_index(drop=True).values.tolist()
    return [' '.join(df['topic']),df['score'].max(),set(df['string']).pop()]
top1('Cruz Roja GTSanto Tomas Castilla')
```

```python
topic_sm_df_gt=pd.DataFrame(list(map(top1,name_list_gt_space)))
```

```python
topic_sm_df_gt.columns=['topic','score','name_sm']
```

```python
topic_sm_df_gt
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 1))
vectorizer.fit(matched_df_gt['name'])
```

```python
topic_match_df_gt=pd.DataFrame(list(map(top1,matched_df_gt['name'])))
```

```python
topic_match_df_gt.columns=['topic','score','name_match']
```

```python
def fuzzy_match(string):
    match=process.extract(string,topic_match_df_gt['topic'],limit=1,scorer=fuzz.token_sort_ratio)[0]
    return [string,match[0]]
```

```python
middle_match_df_gt=pd.DataFrame(list(map(fuzzy_match,topic_sm_df_gt['topic'])))
```

```python
middle_match_df_gt.columns=['sm_topic','name_topic']
```

```python
tmp_df_gt=topic_sm_df_gt.merge(middle_match_df_gt,left_on='topic',how='left',right_on='sm_topic')[["name_sm","sm_topic","name_topic"]].merge(topic_match_df_gt,how='left',left_on='name_topic',right_on='topic')[['name_sm','name_match']].merge(matched_df_gt,how='left',left_on="name_match",right_on='name').drop(columns=['Unnamed: 0','min_dist','name'])
```

```python
pd.concat([infor_gt,tmp_df_gt],axis=1).columns
```

```python
merged_gt=pd.concat([infor_gt,tmp_df_gt],axis=1)[['fb_id', 'latitude', 'longitude','name', 'phone', 'username', 'website', 'single_line_address','name_sm',
       'name_match', 'address', 'web_lat', 'web_lng', 'weight']]
```

```python
merged_gt['distance_page_web']=merged_gt.apply(lambda x:calculate_distance((x['latitude'],x['longitude']),(x['web_lat'],x['web_lng'])),axis=1)
```

```python
merged_gt
```

```python
topic_sm_df_gt=pd.DataFrame(list(map(top1,merged_gt)))
```
